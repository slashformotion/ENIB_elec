# Test LP Filter

Setup

```python
import numpy as np
try:
    from ENIB_lib import HP
except ImportError:
    import sys
    sys.path.append('../.')
    from ENIB_lib import HP
```

Let's create some filters

```python
T1 = HP(1.5, 0.1, 120)
T2 = HP(2, 1.5, 100)
```

Let's create a logspace for later

```python
w = np.logspace(1,3,1000)
```

## Poles / Zeros

```python
for T in [T1,T2]:
    T.pzmap()
```

### Time Response

```python
t = np.arange(0,0.5,0.001)
for T in [T1,T2]:
    T.step(T=t)
```

```python
for index, T in enumerate([T1,T2]):
    print("system {}".format(index))
    print("wp = {} rad/s".format(T.wp))
    print("Tp = {} rad/s".format(T.Tp))
    print("R = {}".format(T.R))
```

### Frequency Response

```python
for T in [T1,T2]:
    T.freqresp(w=w)
```
