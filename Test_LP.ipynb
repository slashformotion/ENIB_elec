# Test LP Filter

```python
import numpy as np
from ENIB_lib import *
```

```python
T1 = LP(2,0.1,100)
T2 = LP(2,2.5,100)
```

## Poles / Zeros

```python
for T in [T1,T2]:
    T.pzmap()
```

### Time Response

```python
t = np.arange(0,0.5,0.001)
for T in [T1,T2]:
    T.step(T=t)
```

```python
for index, T in enumerate([T1,T2]):
    print("system {}".format(index))
    print("wp = {} rad/s".format(T.wp))
    print("Tp = {} rad/s".format(T.Tp))
    print("R = {}".format(T.R))
```

### Frequency Response

```python
w = np.logspace(1,3,1000)
for T in [T1,T2]:
    T.freqresp(w=w)
```
